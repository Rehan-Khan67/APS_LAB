# Shortest Path Using Dijkstra Algorithm in Python
Problem Description:
Given a weighted graph with non negative edge weights. Find the shortest distance from a source vertex to all other vertices using Dijkstra algorithm

#Theory
Dijkstra algorithm finds shortest distance from one source node to all other nodes in a weighted graph with non negative weights. Dictionary structure stores nodes as keys and neighbor nodes with weights as values.

##Algorithm Steps

1. Represent graph using dictionary.
2. Set distance of every node as infinity.
3. Set distance of source node as 0.
4. Select node with smallest distance among unvisited nodes.
5. Update distance of neighbor nodes.
6. Repeat until all nodes get visited.

```python
def dijkstra(graph, start):
    distances = {node: float('inf') for node in graph}
    paths = {node: [] for node in graph}
    distances[start] = 0
    unvisited = list(graph.keys())

    while unvisited:
        curr = min(unvisited, key=lambda n: distances[n])
        unvisited.remove(curr)

        if distances[curr] == float('inf'): break

        for neighbor, weight in graph[curr].items():
            if distances[curr] + weight < distances[neighbor]:
                distances[neighbor] = distances[curr] + weight
                paths[neighbor] = paths[curr] + [curr] # Track path

    return distances, {n: p + [n] for n, p in paths.items() if distances[n] != float('inf')}
```

```python
default_graph = {
    'A': {'B': 4, 'C': 2}, 'B': {'A': 4, 'C': 5, 'D': 10, 'E': 3},
    'C': {'A': 2, 'B': 5, 'D': 3, 'F': 8}, 'D': {'B': 10, 'C': 3, 'E': 4, 'F': 2},
    'E': {'B': 3, 'D': 4, 'F': 6}, 'F': {'C': 8, 'D': 2, 'E': 6}
}

print(f"Default Graph: {default_graph}")
user_input = input("Paste custom graph dict here (or press Enter to use default): ")
graph = eval(user_input) if user_input.strip() else default_graph

start_node = input(f"Enter source node ({', '.join(graph.keys())}): ").strip().upper()

distances, paths = dijkstra(graph, start_node)
print(f"\nShortest paths from {start_node}:")
for node, dist in distances.items():
    path_str = " -> ".join(paths.get(node, []))
    print(f"To {node}: Distance = {dist} | Path: {path_str}")
```
